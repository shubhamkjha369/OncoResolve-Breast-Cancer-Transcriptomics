# Importing Libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path
import joblib

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 20)

import warnings
warnings.filterwarnings("ignore")
```

# Setting Project Directories

```python
CURRENT_DIR = Path.cwd()

PROJECT_ROOT = CURRENT_DIR.parent

DATA_DIR = PROJECT_ROOT / "data"

RAW_DATA_DIR = DATA_DIR / "raw"
PROCESSED_DATA_DIR = DATA_DIR / "processed"
ARTIFACT_DIR = DATA_DIR / "artifacts"

print(PROJECT_ROOT)
print(RAW_DATA_DIR)
```

# Loading Dataset

```python
dataset_path = RAW_DATA_DIR / "Breast_GSE45827.csv"

df = pd.read_csv(dataset_path)
```

```python
float_cols = df.select_dtypes(include=["float64"]).columns

df[float_cols] = df[float_cols].astype(np.float32)

print("Optimization complete.")
```

```python
df = df.drop('samples', axis = 'columns')
```

#### i want to remove the non-cancer/artificial categories (normal, cell_line) and keep only actual cancer subtypes

```python
df = df[~df['type'].isin(['normal', 'cell_line'])]
df = df.reset_index(drop=True)
```

# Exporting Optimized Artifact

```python
parquet_path = PROCESSED_DATA_DIR / "breast_cancer.parquet"

df.to_parquet(parquet_path, index=False)

print(f"Saved to: {parquet_path}")
```
