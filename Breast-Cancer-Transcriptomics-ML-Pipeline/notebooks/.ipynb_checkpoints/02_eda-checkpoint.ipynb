# Importing Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")

plt.style.use("default")
sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns", 100)
```

# Loading Dataset

```python
df = pd.read_parquet(PROCESSED_DATA_DIR / "breast_cancer.parquet")

print("Dataset loaded.")
print("Shape:", df.shape)
```

# Dataset Inspection

```python
df.head()
```

```python
df.info()
```

## Identifying Target Column

```python
target_col = "type"
```

## Identifying Class Distribution

```python
class_counts = df[target_col].value_counts()

print(class_counts)
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x=target_col,
    order=class_counts.index
)

plt.title("Cancer Subtype Distribution")
plt.xticks(rotation=45)
plt.show()
```

## Missing Values

```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

## Separating Features & labels

```python
X = df.drop(columns=[target_col])

y = df[target_col]

print("Features shape:", X.shape)
print("Labels shape:", y.shape)
```

## Gene Statistics

```python
X.describe().T.head()
```

### Gene Expression Distribution

```python
gene_1007_s_at = X.columns[0]

plt.figure(figsize=(8,5))

sns.histplot(X[gene_1007_s_at], bins=30)

plt.title(f"Distribution of {gene_1007_s_at}")
plt.show()
```

### Variance Analysis

```python
gene_variances = X.var().sort_values(ascending=False)

gene_variances.head()
```

```python
plt.figure(figsize=(12,5))

plt.plot(gene_variances.values)

plt.title("Gene Variance Distribution")
plt.xlabel("Genes")
plt.ylabel("Variance")

plt.show()
```

### Correlation Exploration

```python
sample_genes = X.columns[:250]

corr_matrix = X[sample_genes].corr()
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0
)

plt.title("Gene Correlation Heatmap")

plt.show()
```

## Scaling Before PCA

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## PCA

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)
```

```python
print(
    "Explained variance:",
    pca.explained_variance_ratio_
)
```

```python
pca_df = pd.DataFrame({
    "PC1": X_pca[:,0],
    "PC2": X_pca[:,1],
    "Subtype": y
})

plt.figure(figsize=(10,7))

sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="Subtype",
    s=100
)

plt.title("PCA of Breast Cancer Gene Expression")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
# Create dataframe
pca_df = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "PC3": X_pca[:, 2],
    "Subtype": y
})

# 3D Plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# Color mapping
labels = pca_df["Subtype"].unique()
colors = sns.color_palette("Set2", len(labels))

for label, color in zip(labels, colors):
    subset = pca_df[pca_df["Subtype"] == label]

    ax.scatter(
        subset["PC1"],
        subset["PC2"],
        subset["PC3"],
        label=label,
        color=color,
        s=60,
        alpha=0.8
    )

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")

ax.set_title("3D PCA Plot")
ax.legend()

plt.show()
```

# Saving Artifacts

```python
pca_df.to_parquet(
    PROCESSED_DATA_DIR / "pca_2d.parquet",
    index=False
)
print("Done")
```
